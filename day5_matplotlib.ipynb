```python
import pandas as pd
tips=pd.read_csv('Data/tipsdataset.csv')
tips
```

```python
#univariate 
plt.plot(tips['total_bill'])
plt.show()
```

```python
plt.scatter(tips['size'],tips['total_bill'])
plt.title("scatter")
plt.show
```

```python
plt.scatter(tips['size'],tips['total_bill'],s=tips['tip'])
plt.title("scatter")
plt.show
```

```python
plt.pie(
    tips['gender'].value_counts(),
    labels=tips['gender'].value_counts().index,
    autopct='%10.10f%%'
)

plt.show()
```

```python
sizes = tips['gender'].value_counts()
labels = sizes.index
explode = (0.05, 0)  # pehli category ko thoda bahar

fig, ax = plt.subplots()

ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size"
)

plt.show()
```

```python
sns.relplot(
    data=tips,
    x='tip',
    y='size',
    col='gender'
)
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde=True, color='green')
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.histplot(
    data=tips,
    x='total_bill',
    hue='time',
    kde=True,
    palette='Set1'
)

plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True)
sns.histplot(data=tips, x="tip", kde=True)
```

```python
sns.histplot(
    data=tips,
    x='gender',
    hue='time',
    kde=True,
    palette='Set1'
)

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cols = ['gender', 'smoker', 'day', 'time']

for col in cols:
    tips[col] = le.fit_transform(tips[col])
```

```python
tips.corr()
```

```python
sns.heatmap(
tips.corr(),annot=True,fmt='.2f',cmap='viridis',linewidths=0.5,square=True,vmin=-1,vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
