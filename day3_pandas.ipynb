```python
import pandas as pd

# Dictionary se DataFrame banaya
data = {
    "Name": ["Danny", "John", "Sam", "Alex", "Tom"],
    "Age": [20, 25, 22, 24, 21],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

# Pehli 5 rows dikhata hai
print("HEAD:")
print(df.head())

# Aakhri 5 rows dikhata hai
print("\nTAIL:")
print(df.tail())

# Rows aur columns ki count deta hai
print("\nSHAPE:")
print(df.shape)

# DataFrame ki summary deta hai
# Columns, data types, non-null values, memory usage
print("\nINFO:")
df.info()
```
