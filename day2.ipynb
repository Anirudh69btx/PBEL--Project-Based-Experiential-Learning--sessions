```python
# data type
# list
my_list = [1, 2, 3, 4, 5]
# Methods
print(my_list)
my_list.append(6) # Adds an element to the end of the list
print(my_list) # Output: [1, 2, 3, 4, 5, 6]
my_list.insert(2, 10) # Inserts 10 at index 2 print(my_list) # Output: [1, 2, 10, 3, 4, 5, 6]
my_list.remove(3) # Removes the first occurrence of 3
print(my_list) # Output: [1, 2, 10, 4, 5, 6]
my_list.pop() # Removes and returns the last element. print(my_list) # Output: [1, 2, 10, 4, 5]
print(my_list) # Output: [1, 2, 10, 4, 5]
# properties
#ordered and mutable
my_list[2]= 20 # Modifies the element at index 2
print(my_list) # Output: [1, 2, 20, 4, 5]
#heterogeneous & duplicates allowed
zen=[1,2,3,4,5,1,2,3, "ai", True, False]
print(zen) # Output: [1, 2, 3, 4, 5, 1, 2, 3, 'ai', True, False
# properties
```

```python
class Box:
    def __init__(self):
        self.box = []

    def push(self, item):
        self.box.append(item)

    def pop(self):
        return self.box.pop()

    def stack(self):
        print(self.box)


b = Box()

b.push(1)
b.push(2)
b.push(3)

b.stack()

print(b.pop())

b.stack()
```

```python
class Car:

    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year

    def start(self):
        print("Car is starting")

    def stop(self):
        print("Car is stopping")

    def play(self, song):
        print(f"Playing {song}")


bmw = Car("red", "X5", 2020)

print(bmw.color)
print(bmw.model)
print(bmw.year)

bmw.start()
bmw.play("Believer")
bmw.stop()
```

```python
def print_squares(numbers):
    for num in numbers:
        square = num ** 2
        print(f"The square of {num} is {square}")

print_squares([1, 2, 3, 4])
```

```python
def count_up_to(max_number):
    count = 1
    while count <= max_number:
        print(count)
```

```python
def count_to_five():
    i = 1

    while i <= 5:
        print(i)
        i += 1

count_to_five()
```
